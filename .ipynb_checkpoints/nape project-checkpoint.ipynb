## Linear and Polynomial Regression

```python
# Data preparation for models
train_data['Choke Size'].fillna(df_train['Choke Size'].mean(), inplace=True)

X = train_data.drop(['PRODUCTION DATE', 'Oil Production (stb/day)', 
                     'Gas Volume (scf/day)', 'Water Production (stb/day)'], axis=1)
y = train_data['Oil Production (stb/day)']

y = np.reshape(y, (-1, 1))

# Applying normalization in the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)
```

```python
# Handle missing values
# X = pd.DataFrame(X, columns=X.columns).fillna(X.mean())
# y = pd.DataFrame(y).fillna(y.mean())


X_scaled = pd.DataFrame(X_scaled, columns=X.columns).fillna(X.mean())
y_scaled = pd.DataFrame(y_scaled).fillna(y.mean())

# Convert back to numpy arrays 
X_scaled = X_scaled.values

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_lr = lin_reg.predict(X_test)

#y_pred_lr1 = scaler.inverse_transform(y_pred_lr)
#y_test1 = scaler.inverse_transform(y_test)


# Evaluate the model

print('Random Forest Regressor:')
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print('MAE:', mean_absolute_error(y_test, y_pred_lr))
print('R:', r2_score(y_test, y_pred_lr))
```

```python
# A plot of the prediction distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_lr, edgecolors=(0, 0, 0))
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r', lw=4)
plt.xlabel('Actual Oil Production')
plt.ylabel('Predicted Oil Production')
plt.title('MLR')
```

```python
# Polynomial Regression For Oil Production Prediction

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X_scaled)
```

```python
# Splitting the dataset into train 80% and test dataset 20%
X_train, X_test, y_train, y_test  = train_test_split(X_poly, y_scaled, test_size=0.2, 
                                                     random_state=1)
# Fitting model for oil prediction
lin_reg_pol = LinearRegression()
lin_reg_pol.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_lr_pol_oil = lin_reg_pol.predict(X_test)

#y_pred_lr_pol_oil1 = scaler.inverse_transform(y_pred_lr_pol_oil)
#y_test1 = scaler.inverse_transform(y_test)

# Evaluate the model

print('Random Forest Regressor:')
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_lr_pol_oil)))
print('MAE:', mean_absolute_error(y_test, y_pred_lr_pol_oil))
print('R:', r2_score(y_test, y_pred_lr_pol_oil))
```

```python
train_dataset = pd.read_csv('nape_training_data.csv')
#train_dataset['PRODUCTION DATE'] = pd.to_datetime(train_dataset['PRODUCTION DATE'], 
 #                                              infer_datetime_format=True, errors='coerce')
X = train_data.drop(['PRODUCTION DATE', 'Oil Production (stb/day)', 'Gas Volume (scf/day)', 
                     'Water Production (stb/day)'], axis=1)

categorical_columns = X.select_dtypes(include=['object']).columns
X = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

X = X.fillna(X.mean())
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y = train_data['Oil Production (stb/day)']
# Reshaping 1D array to 2D array
y = np.reshape(y, (-1,1))

# Normalizing the target
scaler = StandardScaler()
y_scaled = scaler.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, 
                                                    random_state=42)
```

```python
y_train
```

```python
# Building ANN model
model_ann = Sequential()

model_ann.add(Dense(64, activation='relu'))
model_ann.add(Dropout(0.2)),
model_ann.add(Dense(32, activation='relu'))
model_ann.add(Dropout(0.2)),
model_ann.add(Dense(1))


model_ann.compile(optimizer='adam', loss='mse')

# Train the model
pred = model_ann.fit(X_train, y_train, epochs=400, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# making predictions
y_pred_dl = model_ann.predict(X_test)

#y_pred_dl1 = scaler.inverse_transform(y_pred_dl)
```

```python
y_pred_dl
```

```python
#y_test1 = scaler.inverse_transform(y_test)
#y_pred_dl1 = scaler.inverse_transform(y_pred_dl)

# Model Evaluation
dl_r2_oil = r2_score(y_test, y_pred_dl)
dl_mae_oil = mean_absolute_error(y_test, y_pred_dl)
dl_rmse_oil = np.sqrt(mean_squared_error(y_test, y_pred_dl))


print('MAE: ', dl_mae_oil)
print('MSE: ', dl_rmse_oil)
print('R2 Score: ', dl_r2_oil)
```
